```python
df = pd.read_json("../data/News_Category_Dataset_v2.json", lines=True)
```

```python
df.head()
```

```python
categories = df['category'].unique()
categories
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
def scrape_news(url):
    headers = headers = {
    'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.72 Mobile Safari/537.36'}
    #url = "https://m.huffpost.com/us/entry/us_5b081ab4e4b0802d69caad89"

    r = requests.get(url,headers=headers)
    soup = BeautifulSoup(r.content, 'html')

    contents = soup.select(".entry__text")[0]

    paragraphs = contents.select("p")
    desc = []
    for p in paragraphs:
        desc.append(p.text)
    return " ".join(desc)
```

```python
import time
for category in categories[16:]:
    scrapped_data = {
        "title": [],
        "category": [],
        "desc": []
    }
    print(f"======{category}======")
    temp_df =  df[df['category'] == category]
    count = 0
    for idx, row in temp_df.iterrows():
        if len(scrapped_data['category']) > 500:
            break
        title = row['headline']
        url = row['link']
        try:
            desc = scrape_news(url)
            count += 1
        except Exception:
            continue
        scrapped_data['title'].append(title)
        scrapped_data['category'].append(category)
        scrapped_data['desc'].append(desc)
        print(f"======{count} data Scrapped (INDEX : {idx})======")
        print(f"Title : {title}")
        if count % 5 == 0:
            one_df = pd.DataFrame(scrapped_data)
            one_df.to_csv(f"../data/{category}.csv", index=False)
            print("====SAVED====\n")

        
    one_df = pd.DataFrame(scrapped_data)
    one_df.to_csv(f"../data/{category}.csv", index=False)
    print("====SAVED & SLEEPING====")
    time.sleep(200)
    print("====RESTARTING====")
```
